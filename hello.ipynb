```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parâmetros do gráfico de coração
t = np.linspace(0, 2*np.pi, 100)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# Plotando o gráfico de coração
plt.plot(x, y, 'r-')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Coração')
plt.axis('equal')
plt.show()
```
